```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math as mt
plt.style.use("ggplot")

np.random.seed(44)


def parametric_bootstrap(sample, model, statistic, N):
    arr = []
    n = len(sample)
    for _ in range(N):
        new_sample = model(np.random.uniform(0, 1, n))
        arr.append(statistic(new_sample))

    return np.sort(arr)

def uniform_distribution(x, theta):
    return theta * (x + 1)
```

# Генерируем выборку

```python
n = 100
beta = 0.95
# Пусть 
theta = 10

sample = np.random.uniform(theta, 2 * theta, n)
print("Выборка:", sample)
```

# f) Вычисление доверительных интервалов из пунктов d) и e)

```python
xmax = np.max(sample)

accurate_left = xmax / (1 + ((1 + beta) / 2) ** (1 / n))
accurate_right = xmax / (1 + ((1 - beta) / 2) ** (1 / n))
len_accurate = accurate_right - accurate_left

print("Точный доверительный интервал для theta:", np.round(accurate_left, 4), "< theta <", np.round(accurate_right, 4))
print("Длина точного доверительного интервала для theta:", np.round(len_accurate, 4))
```
