```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, special
import math as mt
plt.style.use("ggplot")
np.random.seed(44)

n = 3
m = 2
alpha = 0.05

x_n = np.array([-1.11, -6.10, 2.42])
y_m = np.array([-2.29, -2.91])

sgm_x_sq = 2
sgm_y_sq = 1
```

```python
x_mean = np.mean(x_n)
y_mean = np.mean(y_m)

delta = (x_mean - y_mean) / np.sqrt(sgm_x_sq / n + sgm_y_sq / m)

U_1alpha = stats.norm.ppf(1 - alpha, loc = 0, scale = 1)

p_value = stats.norm.sf(delta, loc = 0, scale = 1)

print(f"Статистика критерия = {np.round(delta, 5)}, критическая область G: delta >= {np.round(U_1alpha, 5)}")
print("Нет веских оснований отвергнуть H0" if delta < U_1alpha else "Отвергаем H0")

print("Проверка по p-value:")
print(f"Статистика критерия = {np.round(delta, 4)}, p-value = {np.round(p_value, 4)}")
print("Нет веских оснований отвергнуть H0" if p_value > alpha else "Отвергаем H0")
```

```python
figure, axis = plt.subplots(figsize=(16,9))
x = np.linspace(0, 5, 1000)

y = stats.norm.sf(U_1alpha - x / np.sqrt(sgm_x_sq / n + sgm_y_sq / m), loc = 0, scale = 1)

axis.plot(x, y)
axis.set_xlabel("theta", fontsize=20)
axis.set_ylabel("W(theta)", fontsize=20)
axis.set_title("Мощность критерия", fontsize=20)
plt.show()
```
